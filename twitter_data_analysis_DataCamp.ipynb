```python
def count_entries(df, col_name): 
    """Return a dictionary with counts of occurrences as value for each key."""
    langs_count = {} # Initialize an empty dictionary: langs_count
    col = df[col_name] # Extract column from DataFrame: col

    for entry in col: # Iterate over lang column in DataFrame
        if entry in langs_count.keys(): # If the language is in langs_count, add 1 
            langs_count[entry] += 1
        else: # Else add the language to langs_count, set the value to 1
            langs_count[entry] = 1
    return langs_count # Return the langs_count dictionary

# Call count_entries(): result
result = count_entries(tweets_df, 'lang')

# Print the result
print(result)
```
